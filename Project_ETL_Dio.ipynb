# Extract
Extraindo a lista de IDs de usuário a partir do arquivo "Users.xlsx" para cada ID,

```python
import pandas as pd
users = pd.read_excel('Users.xlsx').to_dict(orient='records')


for user in users:
  user['news'] = []
print(users)
```

# Transform
Utilizando a API do OpenAI GPT para gerar uma mensagem de marketing personalizada para cada usuário.


```python
!pip install openai
```

```python
openai_api_key = 'Sua-chave-aqui'
```

```python
from openai import OpenAI

client = OpenAI(api_key=openai_api_key)

def generate_ai_news(user):
  completion = client.chat.completions.create(
    model="gpt-3.5-turbo",
    messages=[
      {
          "role": "system",
          "content": "Você é um especialista em marketing bancário."
      },
      {
          "role": "user",
          "content": f"Crie uma mensagem para {user['Name']} sobre a importância dos investimentos (máximo de 100 caracteres)"
      }
    ]
  )
  return completion.choices[0].message.content.strip('"')

for user in users:
  news = generate_ai_news(user)
  print(news)
```

# Load
Atualizando a lista de "news" de cada usuário no arquivo  com a nova mensagem gerada.

```python
from openai import OpenAI
import pandas as pd


for user in users:
    if 'news' not in user or not isinstance(user['news'], list):
        user['news'] = []

for user in users:
  news_item = generate_ai_news(user)
  user['news'].append(news_item) 
  print(f"Notícia para {user['Name']}: {news_item}")


df_updated = pd.DataFrame(users)


df_updated.to_excel('Users.xlsx', index=False)
print("As notícias foram geradas e salvas em 'Users.xlsx'")
```
